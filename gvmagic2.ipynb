```python
%%dot

digraph{
    fontname="Arial"
    fontsize=20
    label="Fig 1: Pandas Grouping Operations"
    rankdir="LR"
    
    
    // default node properties
    node[shape=box, fontsize=10, fontname="Arial"]

    
    Input [shape="record", label="{{key|A|B|C|A|B|C}|{data|1|2|3|4|5|6}}"]
    Split1 [shape="record", label="{{key|A|A}|{data|1|4}}"]
    Split2 [shape="record", label="{{key|B|B}|{data|2|5}}"] 
    Split3 [shape="record", label="{{key|C|C}|{data|3|6}}"]
    Apply1 [shape="record", label="{{key|A}|{data|5}}"]
    Apply2 [shape="record", label="{{key|B}|{data|7}}"]
    Apply3 [shape="record", label="{{key|C}|{data|9}}"]
    Combine [shape="record", label="{{key|A|B|C}|{data|5|7|9}}"]
    
    subgraph cluster_4{
        label="Data Flow"
        
        subgraph cluster_0{
            penwidth=0
            fontsize=12
            label="Input"
            Input
        }
        subgraph cluster_1{
            penwidth=0
            fontsize=12
            label="Split"
            Split1
            Split2
            Split3
        }
        subgraph cluster_2{
            penwidth=0
            fontsize=12
            label="Apply (sum)"
            Apply1
            Apply2
            Apply3
        }
        subgraph cluster_3{
            penwidth=0
            fontsize=12
            label="Combine"
            Combine
        }
    }
    Input->{Split1, Split2, Split3}
    Split1->Apply1
    Split2->Apply2
    Split3->Apply3
    
    { Apply1, Apply2, Apply3}-> Combine
    
}
```
